# Functions:

## 1. Data type conversion

```python
# Interates through values and if there is a blank instance returns none to avoid an error
# and it convert to the stated conversion type
# if no conversion is needed, just opt-out parse_type
def par_int_float_none(value, **parse_type):
    
    #When the user leaves the parse_type optional argument, the function will ignore it
    if parse_type['parse_type'] != {}:
        parse_type = parse_type['parse_type']['convertype']
    
    #Converts the values provided according to the parse_type
    #Return None for missing blank values
    #Return the same value type for opted-out parse_type
    if value == ' ' or value == '' or value == None:
        return None    
    elif parse_type == 'float':
        return float(value)
    elif parse_type == 'int':
        return int(value)
    else:
        return value
```

```python
def convert_type(list_reader, header, **convertype):
    for rows in list_reader:
        rows[header] = par_int_float_none(rows[header], parse_type = convertype)
```

## 2. Invalid data on Frame & Cleaning

```python
#The formula will create a dataframe with only records with invalid data values for any given columns 
#If the user order the header in a descending order, the records with more issues first, the less memory is used
#If the value is not None under a header, then it will move to the next header to find out validity
def invalid_to_frame(list_reader, *headers, **limit):
    
    #This variable will be the parameter to stop the loop
    #When the user decide to show only a limit of rows
    tolimit = 0
    
    list_values = []
    
    #It will loop vertically trhough the first header
    for rows in list_reader:
        if rows[headers[0]] == None:
            list_values.append(rows)
            tolimit += 1
            
        #It will loop horizontally trhough the next headers, same row and brake
        #Minimize getting in this loop by ordering the headers in a descending order of issues
        else: 
            for i in range(len(headers)):
                if rows[headers[i]] == None:
                    list_values.append(rows)
                    tolimit += 1
                    enter = True
                    break         
 
           
        #Condition will stop the loop at the limit provided
        #It will allow the user to select the number of rows included
        if limit != {}:
            if tolimit >= limit['limit']:
                break

    list_values_df = pd.DataFrame(list_values)
     
    return list_values_df
```

```python
#This formula will create a quick validation table with the difference 
#between the count of records per column and the total record count
def validation_frame(dataframe, key):
    frame = pd.DataFrame(dataframe.count()-dataframe[key].count())
    frame.columns = ['Validation Issues']
    return frame                         
```

```python
#The function will provide only records not inlcuded in the invalid dataset provided
def clean_dataframe(indataframe, frominv_dataset, onkey):
    return indataframe.loc[~indataframe[onkey].isin(frominv_dataset[onkey])]
```

## 3. Proportion Creation from Frame to Numpy

```python
#This function will create a prorportion of each comparing_values by the categorize_value
#Index 0 is the data; index 1 are the keys, on the output array; and index 2 is the category size
from collections import OrderedDict
def ratio_groupby_tonumpy(data_frame, cat_array, what_toassess):
    
    # group data
    success_dat = data_frame.groupby(cat_array)[what_toassess].sum()
    total_dat = data_frame.groupby(cat_array)[what_toassess].count()
    split_count = len(data_frame.groupby(cat_array[len(cat_array)-1])[what_toassess].count())
    
    
    # calculate proportions
    ratio_calc = success_dat / total_dat
    
    
    # convert to numpy array
    np_prop_array = np.split(np.array(ratio_calc), split_count)
    
    # create list_unique
    cat_list = list(OrderedDict.fromkeys(success_dat.keys().get_level_values(0)))
    
    # number of categories
    cat_count = len(data_frame.groupby(cat_array)[what_toassess].count())
    
    # cat_list_full
    if len(cat_array) > 1:
        1
        combine_cat = list(success_dat.keys().get_level_values(0) + " Class " + success_dat.keys().get_level_values(1))
    else:
        combine_cat = 'To get combined categories, use an array with 2 values as the second parameter'
    
    return np_prop_array, cat_list, cat_count, combine_cat
```

## 4. Statistics Detail Data Frame

```python
#Creates a table of details statistics
def stat_detail_frame(data, category_array, what_toassess):
    
    #converting to series
    total = data.groupby(category_array)[what_toassess].count()
    partial = data.groupby(category_array)[what_toassess].sum()
    
    #Getting data into dataframe
    dataframe_conv = pd.DataFrame(total)

    #Creates and renames columns
    dataframe_conv.columns = ['Total Embarked']
    dataframe_conv.loc(axis = 1)['Total Survived'] = partial

    dataframe_conv.loc(axis = 1)['Total Died'] = dataframe_conv.iloc(axis = 0)[:, 0] - dataframe_conv.iloc(axis = 0)[:, 1]

    dataframe_conv.loc(axis = 1)['Survival Ratio'] = (dataframe_conv.iloc(axis = 0)[:, 1] / dataframe_conv.iloc(axis = 0)\
                                                            [:, 0])

    dataframe_conv.loc(axis = 1)['Proportion of Ratio'] = dataframe_conv.iloc(axis = 0)[:, 3] / dataframe_conv.iloc(axis = 0)\
                                                 [:, 3].sum() * 100

    return dataframe_conv
```
